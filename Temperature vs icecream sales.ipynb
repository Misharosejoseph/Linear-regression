```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data: Temperature (°C) vs Ice Cream Sales (in hundreds)
X = np.array([10, 15, 20, 25, 30, 35, 40]).reshape(-1,1)
y = np.array([5, 9, 15, 25, 35, 45, 60])

# Train the model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Visualization
plt.scatter(X, y, color="purple", s=100, label="Actual Sales")
plt.plot(X, y_pred, color="red", linewidth=3, label="Prediction Line")
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales (hundreds)")
plt.title("Temperature vs Ice Cream Sales")
plt.legend()
plt.show()
```
