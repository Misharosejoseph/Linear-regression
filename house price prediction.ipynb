```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data: House Size (in 1000 sq.ft) vs Price (in lakhs)
X = np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1,1)
y = np.array([15, 25, 35, 45, 60, 70, 85])

# Train the model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Visualization
plt.scatter(X, y, color="green", s=100, label="Actual Prices")
plt.plot(X, y_pred, color="orange", linewidth=3, label="Prediction Line")
plt.xlabel("House Size (1000 sq.ft)")
plt.ylabel("Price (Lakhs)")
plt.title("House Price Prediction")
plt.legend()
plt.show()
```
